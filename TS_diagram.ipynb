```python
import numpy as np
from matplotlib import pyplot as plt
import xarray as xr
import gsw
%matplotlib inline 
```

```python
fig,ax=plt.subplots()
fig.set_size_inches(8,7)

s_grid, t_grid = np.meshgrid(
    np.linspace(33.6,34.8,100), #change values for salinity range you want to plot (this is 33.6 to 34.8 psu)
    np.linspace(-1.5,4,100) #change values for temperature range you want to plot (this is -1.5 to 4 degC)
)

pref = 2000 #reference pressure, here this is for sigma2 i.e. reference pressure of 2000 m 
dens_grid = gsw.rho(s_grid.ravel(), t_grid.ravel(), pref).reshape(s_grid.shape)-1000 #calculate sigma2 contours

con = plt.contour(s_grid, t_grid, dens_grid, colors='k') #plot sigma2 contours in black
#ax.plot(AS,CT) #here add the salinity, temperature arrays you want to plot in your TS diagram
plt.clabel(con, fmt='%4.2f',fontsize=14)
plt.xlabel('Salinity [psu]',fontsize=18,fontweight='bold')
plt.ylabel('$\\theta$ [$^{\mathrm{o}}$C]',fontsize=18,fontweight='bold')
plt.tick_params(axis='both', labelsize=16)
plt.xlim([33.6,34.8])
```
